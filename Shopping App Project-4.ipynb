```python
import uuid
import csv

class Categories:
    categories_list = []
    def __init__(self,category_id = uuid.uuid4()):
        category_id = uuid.uuid4()
    @classmethod
    def add_category(cls, category_name,category_id = uuid.uuid4()):
        category = {"id": str(uuid.uuid4()), "name": category_name}
        cls.categories_list.append(category)

    @classmethod
    def delete_category(cls, category_name):
        for category in cls.categories_list:
            if category["name"] == category_name:
                cls.categories_list.remove(category)
                print(f"Category '{category_name}' deleted.")
                return
        print(f"Category '{category_name}' not found.")

    @classmethod
    def save_to_file(cls, filename):
        with open(filename, 'a', newline='') as csvfile:
            fieldnames = ['id', 'name']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            if csvfile.tell() == 0:
                writer.writeheader()
            for category in cls.categories_list:
                writer.writerow(category)
```

```python
import uuid
class Product (Categories):
    products_list = []
    def __init__ (self):
        super().__init__(category_id)
    def __init__ (self,product_name ,product_price ,category_name,product_id = uuid.uuid4()):
        super().__init__(category_name) 
        self.product_name = product_name
        self.product_price = product_price
        self.category_name = category_name
    @classmethod
    def add_product(cls):
        product_name = input("Enter product name: ")
        product_price = float(input("Enter product price: "))
        
        print("Available Categories:")
        for category in Categories.categories_list:
            print(category['name'])
        category_name = input("Choose a category for the product: ")

        newproduct = Product(product_name, product_price,category_name)
        newproduct.products_list.append({
            'product name': newproduct.product_name,
            'product price': newproduct.product_price,
            'category name': newproduct.category_name
        })

        # Open the CSV file in append mode and write the product data
        with open('Products.csv', 'a', newline='') as csvfile:
            fieldnames = ['product name', 'product price','category name']
            writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
            if csvfile.tell() == 0:
                writer.writeheader()
            writer.writerow({
                'product name': newproduct.product_name,
                'product price': newproduct.product_price,
                'category name': newproduct.category_name
            })
    @classmethod
    def load_products_from_csv(cls, category_name):
        products = []
        with open('Products.csv', 'r') as csvfile:
            reader = csv.DictReader(csvfile)
            for row in reader:
                if row['category name'] == category_name:
                    products.append({
                        'product name': row['product name'],
                        'product price': float(row['product price'])
                    })
        return products       
   
    @classmethod
    def display_products_by_category(cls, category_name):
        products = cls.load_products_from_csv(category_name)
        for product in products:
            print(f"Name: {product['product name']}, Price: {product['product price']}")

    @classmethod
    def add_to_cart(cls, username, productName):
        selected_product = None
        for product in cls.products_list:
            if product['product name'] == productName:
                selected_product = product
                break

        if selected_product:
            with open(f'{username}_cart.csv', 'a', newline='') as csvfile:
                fieldnames = ['product name', 'product price']
                writer = csv.DictWriter(csvfile, fieldnames=fieldnames)
                if csvfile.tell() == 0:
                    writer.writeheader()
                writer.writerow({
                    'product name': selected_product['product name'],
                    'product price': selected_product['product price']
                })
            print("Product added to the cart successfully!")
        else:
            print("Invalid product name.")
    @classmethod
    def user_interaction(cls, username):
        while True:
            print("1. Choose a category")
            print("2. Display products")
            print("3. Add product to cart and save cart")
            print("4. Exit")
            choice = input("Enter your choice: ")

            if choice == '1':
                print("Available Categories:")
                for category in Categories.categories_list:
                    print(category['name'])
                category_name = input("Enter category name: ")
                cls.display_products_by_category(category_name)

                product_name = input("Enter product name: ")
                product_available = False
                for product in cls.products_list:
                    if product['product name'] == product_name:
                        print("Product available")
                        product_available = True
                        break
                if not product_available:
                    print("Product unavailable")

            elif choice == '2':
                print("Displaying all products:")
                cls.display_products_by_category(category_name)
            elif choice == '3':
                productName = input("Enter the product name to add to your cart: ")
                payment_method = input("Choose a payment method (e.g., credit card, PayPal, etc.): ")
       
                cls.add_to_cart(username, productName)
            elif choice == '4':
                print("Exit done.")
                break
            else:
                print("Invalid choice. Please try again.")
```
